```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.savefig("../../img/03_matplotlib/01.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
plt.savefig("../../img/03_matplotlib/02.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], color='red', linewidth=2, linestyle='--', marker='o')
plt.savefig("../../img/03_matplotlib/03.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.plot([1, 2, 3], [1, 4, 9], 'ro--')  # red, circles, dashed
plt.savefig("../../img/03_matplotlib/04.png", dpi=300, bbox_inches="tight")
plt.show()
```
