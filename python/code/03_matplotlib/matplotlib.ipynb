```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.savefig("../../img/03_matplotlib/01.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
plt.savefig("../../img/03_matplotlib/02.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], color='red', linewidth=2, linestyle='--', marker='o')
plt.savefig("../../img/03_matplotlib/03.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.plot([1, 2, 3], [1, 4, 9], 'ro--')  # red, circles, dashed
plt.savefig("../../img/03_matplotlib/04.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.plot([1, 2, 3], [1, 4, 9], label='Line 1')
plt.plot([1, 2, 3], [2, 5, 10], label='Line 2')
plt.legend()
plt.savefig("../../img/03_matplotlib/05.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]
plt.plot(x, y1, 'r-', x, y2, 'b--')
plt.savefig("../../img/03_matplotlib/06.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.scatter([1, 2, 3, 4], [1, 4, 9, 16])
plt.savefig("../../img/03_matplotlib/07.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
sizes = [20, 50, 100, 200]
colors = [10, 20, 30, 40]
plt.scatter([1, 2, 3, 4], [1, 4, 9, 16], s=sizes, c=colors, cmap='viridis', alpha=0.6)
plt.colorbar()  # Show color scale
plt.savefig("../../img/03_matplotlib/08.png", dpi=300, bbox_inches="tight")
plt.show()
```
